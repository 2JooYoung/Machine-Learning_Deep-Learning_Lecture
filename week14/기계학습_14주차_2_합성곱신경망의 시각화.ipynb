## 14장 합성곱 신경망의 시각화

### 1. 가중치 시각화

```python
from tensorflow import keras
```

```python
# 앞서 만든 모델이 어떤 가중치를 학습했는지 확인하기 위해 체크포인트 파일을 읽어 들인다.
model = keras.models.load_model('best-cnn-model.h5')
```

```python
# 모델에 추가한 층은 layers 속성에 저장되어 있다. 이 속성은 파이썬 리스트이다.
model.layers
```

```python
# layer 속성의 첫번째 원소 layers[0]은 weights 속성이다.
# weights 속성의 첫번째 원소 weights[0]은 가중치이고, 두번째 원소 weights[1]은 절편이다.
conv = model.layers[0]

print(conv.weights[0].shape, conv.weights[1].shape)
```

```python
# 가중치를 numpy()를 사용해 넘파이 배열로 변환한다.
conv_weights = conv.weights[0].numpy()

# 가중치 배열의 평균과 표준편차를 구한다.
print(conv_weights.mean(), conv_weights.std())
```

```python
# 가중치가 어떤 분포를 가졌는지 직관적으로 이해하기 쉽도록 히스토그램을 그린다.
import matplotlib.pyplot as plt

# hist()는 히스토그램을 그리기 위해 1차원배열로 전달해야한다.
# reshape()으로 1개의 열이 있는 배열로 변환한다.
plt.hist(conv_weights.reshape(-1, 1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

```python
# 32개의 커널을 16개씩 두 줄에 그림으로 출력한다.
fig, axs = plt.subplots(2, 16, figsize=(15,2))

# (3, 3, 1, 32)에서 마지막 차원이 필터의 개수이며 마지막 차원을 순회하면서 그림을 출력한다.
# vmin과 vmax으로 컬러맵의 표현 범위를 지정한다.
for i in range(2):
  for j in range(16):
    axs[i, j].imshow(conv_weights[:,:,0,i*16 + j], vmin=-0.5, vmax=0.5)
    axs[i, j].axis('off')

plt.show()
```

### 2. 훈련하지 않은 가중치 시각화

```python
# 훈련하지 않은 빈 합성곱 신경망을 만들어서 가중치를 시각화 해본다.
no_training_model = keras.Sequential()

no_training_model.add(keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same', input_shape=(28,28,1)))
```

```python
# 모델의 layers[0]은 weights 속성이다.
no_training_conv = no_training_model.layers[0]

# weights[0]은 가중치이고 weights[1]은 절편이다.
print(no_training_conv.weights[0].shape)
```

```python
# 가중치를 배열로 만들어서 평균과 표준편차를 구한다.
no_training_weights = no_training_conv.weights[0].numpy()

print(no_training_weights.mean(), no_training_weights.std())
```

```python
# 훈련되지 않은 가중치를 히스토그램으로 그린다.
plt.hist(no_training_weights.reshape(-1,1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

```python
# 훈련되지 않은 가중치를 그림으로 출력한다.
fig, axs = plt.subplots(2, 16, figsize=(15,2))

for i in range(2):
  for j in range(16):
    axs[i, j].imshow(no_training_weights[:,:,0,i*16+j], vmin=-0.5, vmax=0.5)
    axs[i, j].axis('off')

plt.show()
```

### 3. 함수형 API

```python
# 이미 만들어진 신경망 모델(model)의 layers 속성을 확인한다.
model.layers
```

```python
# 신경망 모델의 input 속성을 확인한다.
print(model.input)
```

```python
# 함수형API는 케라스의 Model클래스를 사용하여 모델을 만든다.
# Model 클래스는 기존 신경망에서 특정 층을 분리해내서 모델을 만들 수 있다.

conv_acti = keras.models.Model(model.input, model.layers[0].output)
```

### 4. 특성 맵 시각화

```python
# 패션 MNIST 데이터셋을 로드한다.
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
# 훈련세트에 있는 첫번째 샘플을 그림으로 출력한다.
plt.imshow(train_input[0], cmap='gray_r')
plt.show()
```

```python
# 첫번째 샘플을 (28, 28, 1) 크기로 변경하고 전처리를 한다.
inputs = train_input[0:1].reshape(-1, 28, 28, 1)/255.0

# 첫번째 샘플을 conv_acti 모델로 예측하여 특성맵을 반환한다.
feature_maps = conv_acti.predict(inputs)
```

```python
# 특성맵의 크기를 확인한다.
# 첫번째 차원은 배치차원이며 샘플을 하나 입력했기 때문에 1이 된다.
print(feature_maps.shape)
```

```python
# 32개의 커널을 4행 8열의 그림으로 출력한다.
fig, axs = plt.subplots(4, 8, figsize=(15,8))

# (1, 28, 28, 32)에서 마지막 차원이 필터의 개수이며 마지막 차원을 순회하면서 그림을 출력한다.
# vmin과 vmax으로 컬러맵의 표현 범위를 지정한다.
for i in range(4):
  for j in range(8):
    axs[i, j].imshow(feature_maps[0,:,:,i*8 + j])
    axs[i, j].axis('off')

plt.show()
```
